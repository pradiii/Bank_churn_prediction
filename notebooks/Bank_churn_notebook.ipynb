```python
train_df = train_raw[['CreditScore', 'Geography', 'Gender','Age', 'Tenure', 'Balance', 'NumOfProducts',
                      'HasCrCard','IsActiveMember', 'EstimatedSalary', 'Exited']]
train_df.head()
```

```python
cat_column_one_hot_encode = ['Geography'] 
cat_column_l_encode = ['Gender']
num_columns = ['CreditScore','Age', 'Tenure', 'Balance', 'NumOfProducts',
                      'HasCrCard','IsActiveMember', 'EstimatedSalary', 'Exited']
```

```python
train_df.info()
```

```python
train_df.Geography.value_counts()
```

```python
#Lable encoding on 'Gender' Column.
for col1 in cat_column_l_encode:
    l_encoder =  LabelEncoder().fit(train_df[cat_column_l_encode])
    train_df[col1]  = l_encoder.transform(train_df[cat_column_l_encode])
```

```python
train_df.head()
```

```python
#One encoding on 'Geography' Column.
one_encode = OneHotEncoder(sparse_output=False)
encode_vals = one_encode.fit_transform(train_df[cat_column_one_hot_encode])
encoded_df = pd.DataFrame(encode_vals,columns=one_encode.get_feature_names_out(cat_column_one_hot_encode))
```

```python
encoded_df.head()
```

```python
train_df1 = pd.concat([train_df,encoded_df],axis=1)
train_df1.head()
```

```python
train_df1.drop('Geography',axis=1,inplace=True)
```

```python
train_df1.head()
```

```python
min_max_scale_cols = ['CreditScore','Age','Tenure','Balance','NumOfProducts','EstimatedSalary']
```

```python
min_max = MinMaxScaler()
train_df1[min_max_scale_cols]  = min_max.fit_transform(train_df1[min_max_scale_cols])
```

```python
train_df1.head()
```

```python
train_df1.describe()
```

```python
#Split the data into dependent and indepedent variables/ target and input variables.
X = train_df1.drop('Exited',axis=1)
y = train_df1['Exited']
```

```python
#Split the data into training and testing
train_X,test_X,train_y,test_y = train_test_split(X,y,train_size=0.7)
```

```python
train_X.shape,train_y.shape
```

```python
lg = LogisticRegression()
lg.fit(train_X,train_y)
predictions =  lg.predict(train_X)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
print(classification_report(train_y,predictions))
```
